```python
import sqlite3

# Establecer una conexión a la base de datos SQLite (creará la base de datos si no existe)
conn = sqlite3.connect("dbcheckdemands.db")

# Crear un objeto cursor para ejecutar comandos SQL
cursor = conn.cursor()

# Definir el script SQL para crear las tablas
script_sql = """
CREATE TABLE IF NOT EXISTS `request` (
  `id_request` INTEGER PRIMARY KEY,
  `fk_id_institucion` INTEGER,
  `status` TEXT CHECK (status IN ('active', 'inactive', 'error')),
  `id_filed` INTEGER,
  `fk_id_institution` INTEGER,
  `fk_id_ws_samai` INTEGER,
  `fk_id_ws_judicial` INTEGER,
  `fk_id_location` INTEGER,
  `update_date` DATETIME,
  `added_at` DATETIME,
  `modified_at` DATETIME
);

CREATE TABLE IF NOT EXISTS `ws_samai` (
  `id_ws_samai` INTEGER PRIMARY KEY,
  `status` TEXT CHECK (status IN ('update', 'outdated', 'error', 'inactive')),
  `update_date` DATETIME,
  `added_at` DATETIME,
  `modified_at` DATETIME
);

CREATE TABLE IF NOT EXISTS `ws_judicial` (
  `id_ws_judicial` INTEGER PRIMARY KEY,
  `status` TEXT CHECK (status IN ('update', 'outdated', 'error', 'inactive')),
  `update_date` DATETIME,
  `added_at` DATETIME,
  `modified_at` DATETIME
);

CREATE TABLE IF NOT EXISTS `institution` (
  `id_institution` INTEGER PRIMARY KEY,
  `name` TEXT,
  `status` TEXT CHECK (status IN ('active', 'inactive')),
  `id_samui` INTEGER,
  `id_juridica` INTEGER
  `fk_id_location` INTEGER
);

CREATE TABLE IF NOT EXISTS `location` (
  `id_location` INTEGER PRIMARY KEY,
  `name` TEXT,
  `type` TEXT CHECK (type IN ('country', 'department', 'region', 'city')),
  `fk_id_location` INTEGER,
  `added_at` DATETIME,
  `modified_at` DATETIME
);

CREATE TABLE IF NOT EXISTS `actions` (
  `id_action` INTEGER PRIMARY KEY,
  `fk_id_request` INTEGER,
  `id_filed` TEXT,
  `registration_date` TEXT,
  `update_date` DATETIME
  `action` TEXT,
  `detail` TEXT,
  `status` TEXT,
  `annexes`  TEXT,
  `index` INT
);

CREATE TABLE IF NOT EXISTS `email` (
  `id_email` INTEGER PRIMARY KEY,
  `fk_id_request` INTEGER,
  `status` TEXT CHECK (status IN ('verified', 'not verified')),
  `added_at` DATETIME
);

ALTER TABLE `actions` ADD FOREIGN KEY (`fk_id_request`) REFERENCES `request` (`id_request`);

ALTER TABLE `request` ADD FOREIGN KEY (`fk_id_institucion`) REFERENCES `institution` (`id_institution`);

ALTER TABLE `ws_samai` ADD FOREIGN KEY (`id_ws_samai`) REFERENCES `request` (`fk_id_ws_samai`);

ALTER TABLE `ws_judicial` ADD FOREIGN KEY (`id_ws_judicial`) REFERENCES `request` (`fk_id_ws_judicial`);

ALTER TABLE `request` ADD FOREIGN KEY (`fk_id_location`) REFERENCES `location` (`id_location`);

ALTER TABLE `location` ADD FOREIGN KEY (`fk_id_location`) REFERENCES `location` (`id_location`);

ALTER TABLE `email` ADD FOREIGN KEY (`fk_id_request`) REFERENCES `request` (`id_request`);
"""

# Ejecutar el script SQL para crear las tablas y relaciones
cursor.executescript(script_sql)

# Guardar los cambios en la base de datos
conn.commit()

# Cerrar la conexión a la base de datos
conn.close()

print("Base de datos creada exitosamente.")
```
